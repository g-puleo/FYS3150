```python
#THIS CODE PLOTS THE HISTOGRAMS FOR THE DISTRIBUTION OF THE ENERGY PER SPIN.
#NOTE THAT YOU MIGHT NEED TO HAVE LATEX INSTALLED ON YOUR PC
import matplotlib
matplotlib.rcParams['text.usetex'] = True
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

L = 20
T = [1.0, 2.4]
binwidth = 0.01

data1 = pd.read_csv("LatticeData_20_" + str(T[0]) + "_D.csv")	
epsilons = data1["energy"]/(L*L) 
N = int((max(epsilons[25000 : -1])-min(epsilons[25000:-1]))/binwidth)
variance = np.var(epsilons[25000 : -1])
mean = np.mean(epsilons[25000 : -1])
n, bins, patches = plt.hist(epsilons[25000:-1], bins = N, density=True, facecolor='green', alpha=0.75)

plt.xlabel('$\epsilon$  ($J/$spin)', fontsize=16)
plt.ylabel('Probability density', fontsize=16)
texstr = '\n'.join((r'$\mu = %.4f$ $J/$spin'%(mean, ), 
	 '$\sigma^2=%.4f$ $(J/$spin)$^2$'%(variance, )))
props = dict(boxstyle='round', facecolor='white', alpha=1.0)

# place a text box in upper left in axes coords
plt.text(-1.94, 90, texstr, fontsize=14,
        verticalalignment='top', bbox=props)
plt.grid(True) 
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
#plt.axis([min(epsilons[25000:-1]), max(epsilons[25000:-1]), 0, 100])
plt.yscale('log')
plt.savefig("1.0_histogram.pdf", dpi=300, format="pdf", bbox_inches='tight')
plt.show()

data24 = pd.read_csv("LatticeData_20_" + str(T[1]) + "_D.csv")	
epsilons24 = data24["energy"]/(L*L) 
N = int((max(epsilons24[25000 : -1])-min(epsilons24[25000:-1]))/binwidth)
variance24 = np.var(epsilons24[25000 : -1])
mean24 = np.mean(epsilons24[25000 : -1])
n, bins, patches = plt.hist(epsilons24[25000:-1], bins = N, density=True, facecolor='green', alpha=0.75)

plt.xlabel('$\epsilon$ ($J/$spin)', fontsize=16)
plt.ylabel('Probability density', fontsize=16)
texstr = '\n'.join((r'$\mu = %.4f$ $J/$spin)'%(mean24, ), 
	 '$\sigma^2=%.4f$ $(J/$spin)$^2$'%(variance24, )))
props = dict(boxstyle='round', facecolor='white', alpha=1.0)

# place a text box in upper left in axes coords
plt.text(-1.8, 2.7, texstr, fontsize=14,
        verticalalignment='top', bbox=props)
plt.grid(True) 
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.axis([min(epsilons24[25000:-1]), max(epsilons24[25000:-1]), 0, (max(epsilons24[25000:-1])-min(epsilons24[25000:-1]))/np.sqrt(np.sqrt(variance24))])
plt.savefig("2.4_histogram.pdf", format="pdf", dpi=300, bbox_inches='tight')
plt.show()
	 

```
